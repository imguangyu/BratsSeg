# Inspect the dataset

```python
import numpy as np
```

```python
train = np.load('brats_112/train.npz', allow_pickle=True)
```

```python
imgs = train['images']
labels = train['labels']
```

Modalities: ['flair', 't1', 't2']

classes: Â 
-           0: Background
-           1: Necrotic and non-enhancing tumor core
-           2: Peritumoral edema
-           3: Enhancing tumor

```python
import matplotlib.pyplot as plt

example = imgs.item()['flair'][0]
label = labels[0]

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(example, cmap='gray')
plt.title('Flair Image')
plt.subplot(1, 2, 2)
plt.imshow(label, cmap='gray')
plt.title('Label')
plt.show()



```
